```python
from ultralytics import YOLO

yolo = YOLO("./yolov8n.pt",task="detect")

# conf表示的是置信度
result = yolo(source="./ultralytics/assets/bus.jpg",save=True,conf=0.05)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#　[:,:,::-1] rgb
plt.imshow(result[0].plot()[:,:,::-1])
```
